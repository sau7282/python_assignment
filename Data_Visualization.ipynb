```python
# 5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using 
# a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
n_samples = 100
x1 = np.random.normal(0, 1, n_samples)
x2 = x1 + np.random.normal(0, 0.5, n_samples)
x3 = x2 + np.random.normal(0, 0.5, n_samples)
x4 = np.random.normal(0, 1, n_samples)
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
x = np.linspace(0, 10, 100) 
y = np.sin(x) 
output_notebook()  
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2, line_color='blue')
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for 
# the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
np.random.seed(42)  
n = 50 
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(10, 100, size=n)
colors = np.random.rand(n, 3) 
colors_hex = [f'#{int(r*255):02x}{int(g*255):02x}{int(b*255):02x}' for r, g, b in colors]
output_notebook()
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.scatter(x, y, size=sizes, color=colors_hex, alpha=0.6, legend_label="Random Points")
show(p)
```

```python
# 3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
output_notebook() 
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9, color=factor_cmap('x', palette='Spectral11', factors=fruits))
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_notebook()
p = figure(title="Histogram of Normally Distributed Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency',
           tools='reset,save',
           height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10) 
# y= np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.layouts import gridplot
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10) 
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
data = []
for i in range(data_heatmap.shape[0]):
    for j in range(data_heatmap.shape[1]):
        data.append((xx[i, j], yy[i, j], data_heatmap[i, j]))
p = figure(title='Heatmap', x_axis_label='X', y_axis_label='Y', 
           x_range=(0, 1), y_range=(0, 1), 
           toolbar_location=None)
source = {'x': [d[0] for d in data], 
          'y': [d[1] for d in data], 
          'value': [d[2] for d in data]}
mapper = LinearColorMapper(palette='Viridis256', low=min(source['value']), high=max(source['value']))
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=linear_cmap('value', palette='Viridis256', low=min(source['value']), high=max(source['value'])),
        line_color=None)
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
output_notebook()
show(p)

```
