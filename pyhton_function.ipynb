```python
''''
In Python, both functions and methods are callable blocks of code that perform a specific task, 
but they have key differences in how they are defined and used:

1. Function->A function is a standalone block of code that performs a specific task and can be called independently.
It is defined using the def keyword.
A function can exist outside of any class and can be called using its name directly.
Example:
def my_function():
    print("This is a function")
my_function()  # Calling a function

2. Method->A method is a function that is associated with an object (instance) of a class. 
It is defined within a class and can only be called on instances of that class.
Methods are implicitly passed the instance (commonly referred to as self) as their first argument.
It is also defined using the def keyword but must be accessed through an object or class.
Example:
class MyClass:
    def my_method(self):
        print("This is a method")

obj = MyClass()
obj.my_method()  # Calling a method via an object

Function: Independent of classes and can be called directly.
Method: Defined inside a class and called on an object (or class instance).Methods implicitly take the 
object itself (self) as their first argument.
'''
```

```python
'''
In Python, arguments and parameters are terms used to refer to values that are passed into functions and variables
that hold those values.

1. Parameters->Parameters are the variables listed inside the parentheses in the function definition.
They act as placeholders for the values that the function will receive when it is called.
A function can have zero, one, or multiple parameters.
Example:
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

2. Arguments->Arguments are the actual values passed to the function when it is called. 
These values are assigned to the corresponding parameters defined in the function.
Example:
greet("Alice")  # "Alice" is an argument
In the above example, name is the parameter, and "Alice" is the argument.

Types of Function Arguments in Python:
    
1.Positional Arguments->These are arguments that are passed to a function in the order in which the 
parameters are defined.
Example:
def add(a, b):
    return a + b
add(5, 3)  # Positional arguments: 5 is assigned to 'a', 3 is assigned to 'b'

2.Keyword Arguments->These are arguments that are passed by explicitly naming the parameter in the function call, 
allowing arguments to be passed in any order.
Example:
def add(a, b):
    return a + b

add(b=3, a=5)  # Keyword arguments: specifying which value goes to which parameter

3.Default Arguments->Parameters can have default values, which are used if no argument is passed for them 
during the function call.
Example:
def greet(name="Stranger"):
    print(f"Hello, {name}!")
greet()  # No argument is passed, so the default value "Stranger" is used
greet("Alice")  # Argument is passed, so "Alice" is used instead

4.Arbitrary Arguments (*args)->If you want to pass a variable number of arguments to a function, 
you can use *args. This allows the function to accept any number of positional arguments.
Example:
def add_all(*args):
    return sum(args)

add_all(1, 2, 3, 4)  # Can pass any number of arguments

Parameters: Variables in the function definition.
Arguments: Actual values passed to the function when it is called.
'''
```

```python
'''
In Python, functions can be defined and called in several ways, providing flexibility for handling 
different programming needs.

1. Normal Function Definition->The most common way to define a function is by using the def keyword, 
followed by the function name and parameters (if any).

Syntax:
def function_name(parameters):
    # Function body
    return value

Example:
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))

output: Hello Alice


2. Function with Default Parameters->We can define functions with default parameter values. 
If no argument is passed for the parameter during the function call, the default value is used.

Syntax:
def function_name(parameter=value):
    # Function body
    return value

Example:
def greet(name="Stranger"):
    return f"Hello, {name}!"

print(greet())       
Output: Hello Stranger
print(greet("Alice")) 
Output: Hello Alice
    
3. Lambda (Anonymous) Functions->A lambda function is a small anonymous function defined using the lambda keyword.
It can have any number of arguments but only one expression.

Syntax:
lambda arguments: expression
Example:
# Lambda function definition
add = lambda x, y: x + y
print(add(5, 3)) 
Output: 8

4. Recursive Functions->A recursive function is one that calls itself within its own definition. 
It must have a base case to terminate the recursion.

Syntax:
def function_name(parameters):
    # Base case
    if condition:
        return value
    # Recursive case
    return function_name(modified_parameters)
Example:
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))
Output: 120

5. Function with *args (Variable Number of Positional Arguments)->We can define a function to accept a variable 
number of positional arguments using *args.

Syntax:
def function_name(*args):
    # Function body
    return value
Example:
def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3))
Output: 6

6. Function with **kwargs (Variable Number of Keyword Arguments)->We can define a function to accept a variable 
number of keyword arguments using **kwargs.

Syntax:
def function_name(**kwargs):
    # Function body
    return value
Example:
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Alice", age=30, city="New York")

7. Nested Functions->A function can be defined inside another function. These are called nested functions.

Syntax:
def outer_function():
    def inner_function():
        # Inner function body
        return value
    return inner_function()
Example:
def outer_function():
    def inner_function():
        return "Hello from the inner function!"
    
    return inner_function()

print(outer_function())

10. Calling Methods (Object-Oriented Approach)->In object-oriented programming, we can define functions inside 
classes (methods), and they are called using class instances.

Example:
class MyClass:
    def my_method(self):
        return "This is a method!"

obj = MyClass()
print(obj.my_method())

Output:This is a method!
'''
```

```python
'''
In Python, iterators and iterables are two related concepts used for looping or iterating over data structures, 
but they have distinct roles and characteristics.

1. Iterable->An iterable is any Python object capable of returning its members one at a time. 
It can be passed to the iter() function to get an iterator. An object is considered iterable if it 
implements the __iter__() method, which returns an iterator. 
Examples of iterables include data structures like lists, tuples, strings, dictionaries, sets, and 
even file objects.

Example:
Lists: [1, 2, 3]
Strings: "Hello"
Sets: {1, 2, 3}

Example:
my_list = [1, 2, 3] 
for item in my_list:
    print(item) 
    
2. Iterator->An iterator is an object that represents a stream of data, and it can be used to traverse through 
all the values of an iterable, one by one.
An iterator is an object that implements two methods:
__iter__() – This method returns the iterator object itself.
__next__() – This method returns the next element from the iterator. When there are no more elements to return, it raises a StopIteration exception to signal that the iteration has finished.

Examples:
We can convert an iterable (like a list) into an iterator by using the iter() function.

Example:
my_list = [1, 2, 3]  
iterator = iter(my_list) 

print(next(iterator))  
Output: 1
print(next(iterator)) 
Output: 2
print(next(iterator)) 
Output: 3

Calling next() again will raise a StopIteration exception

Iterable->An object that can return an iterator (like lists, tuples, etc.) and can be looped over in a for loop.
Iterator->An object that represents a stream of data, produced by calling iter() on an iterable, 
and accessed via next() until the StopIteration exception is raised.Iterators allow lazy evaluation
(i.e., producing values one at a time, only when needed).
'''
```

```python
'''
In Python, generators are a special type of iterator that allow us to iterate over a sequence of values lazily,
meaning they produce items one at a time and only when requested. Generators are often used for large datasets or 
infinite sequences because they generate values on the fly rather than storing everything in memory at once.

Lazy Evaluation->Generators do not compute or store all their values upfront. Instead, they generate values 
one at a time when needed, which makes them memory-efficient, especially when working with large datasets.

yield Keyword->Generators are defined using a function that contains one or more yield statements instead of 
return. The yield statement pauses the function, saving its state, and returns the next value
when the generator is iterated over. When yield is called again, the function resumes execution from where 
it left off.

Iteration->We can iterate over a generator using a for loop or manually with next(). 
Once all values have been generated, the generator raises a StopIteration exception, signaling that the iteration
has finished.


A generator function is defined like a normal function but uses yield to return values one at a time.

Example:
def my_generator():
    yield 1  
    yield 2  
    yield 3  

gen = my_generator()
print(next(gen))  
Output: 1
print(next(gen)) 
Output: 2
print(next(gen)) 
Output: 3

Calling next() again will raise StopIteration

2. Generator Expression (similar to list comprehension)->We can also create a generator using a generator 
expression, which is similar to list comprehension but uses parentheses () instead of square brackets [].

Example:
gen_exp = (x**2 for x in range(5))
for value in gen_exp:
    print(value) 

Output: 0, 1, 4, 9, 16


Generator Function->Uses yield to return a value, but pauses and retains its state, allowing it to resume where it 
left off.It can produce multiple values over time.

Example:
def generator_function():
    yield 1
    yield 2
    
Advantages of Generators:

Memory Efficiency->Since generators generate values on the fly, they are much more memory-efficient for 
large datasets compared to lists, which store all elements in memory at once.
Performance:
For large or infinite data streams, generators are faster because they compute values only when needed.
State Persistence:

Generators can maintain local variables between successive yield calls, making them ideal for tasks like reading large files line-by-line.
Simpler Code:

Generators often result in cleaner and more readable code for iterating over sequences.

Generators are special iterators in Python that generate values lazily using the yield keyword.
They are memory-efficient, particularly useful for large datasets or infinite sequences.
Generator functions use yield instead of return, allowing them to produce multiple values over time.
Generator expressions provide a concise way to create generators without using full function syntax.

'''
```

```python
# 7.What are the advantages of using generators over regular functions?
```

```python
'''
Generators offer several advantages over regular functions, especially when dealing with large datasets, 
infinite sequences, or scenarios where memory efficiency and performance are important. 

1. Memory Efficiency (Lazy Evaluation)->Generators produce values one at a time and only when needed, 
rather than storing all values in memory at once. This is called lazy evaluation. In contrast, regular functions 
that return collections (like lists or tuples) allocate memory for all the elements upfront, 
which can be inefficient for large datasets.

Example:
def count_up_to(n):
    for i in range(n):
        yield i

def count_up_to_list(n):
    return list(range(n))
    
If n is a large number, the generator is much more memory-efficient because it yields values one by one, 
while the regular function returns a complete list, which consumes a lot of memory.

2. Performance and Speed->Generators are generally faster for large or infinite sequences because they compute and 
yield values as needed, rather than constructing the entire result at once. This also reduces initialization 
time when handling large datasets.

Example: 
If you only need the first 10 items of a large sequence, a generator can start yielding those items 
immediately, without waiting to generate the entire sequence.

3. Suitability for Infinite Sequences->Generators can handle infinite sequences efficiently because they do not
need to store all values in memory. Regular functions, on the other hand, cannot return infinite sequences 
since they would attempt to store all values, which is impossible.

Example: A generator that produces Fibonacci numbers infinitely:

def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
        
A regular function would run into memory or performance issues if tasked with returning an infinite sequence.

4. State Persistence Without Complexity->Generators automatically save their state between calls to next() 
(or between iterations in a for loop), making them ideal for tasks like reading large files line by line. 
This allows for efficient handling of ongoing processes without the complexity of managing state manually.

Example: Reading a large file line by line using a generator:

def read_large_file(file_name):
    with open(file_name) as file:
        for line in file:
            yield line
A regular function would have to load the entire file into memory at once or maintain complex state.

5. Cleaner and More Readable Code->Generators often make the code more concise and readable, especially for 
sequences or iterations where the result is produced progressively. The use of yield in place of explicit state
management or large data structures simplifies the logic of the function.

Example:
Generator-based Fibonacci sequence:
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
        
Regular function equivalent:
def fibonacci(n):
    result = []
    a, b = 0, 1
    while len(result) < n:
        result.append(a)
        a, b = b, a + b
    return result
    
The generator version is simpler and avoids managing a list manually.

6. Pipelining Operations (Composability)->Generators allow for chaining operations in a pipeline, where each step
processes a stream of data lazily. This enables efficient handling of large data flows where each step in the 
pipeline only processes one element at a time, reducing memory overhead.

Example:
def count_up_to(n):
    for i in range(n):
        yield i

def square(nums):
    for num in nums:
        yield num ** 2

def filter_even(nums):
    for num in nums:
        if num % 2 == 0:
            yield num

nums = count_up_to(10)
squares = square(nums)
even_squares = filter_even(squares)

for val in even_squares:
    print(val)  
Output: 0, 4, 16, 36, 64
each function processes the data lazily, passing it along the pipeline without creating intermediate data 
structures.
'''
```

```python
# 8.What is a lambda function in Python and when is it typically used?
```

```python
'''
A lambda function in Python is a small, anonymous function defined with the lambda keyword instead of the regular 
def keyword. Lambda functions are often used for short, simple operations that are needed only temporarily and 
are typically passed as arguments to higher-order functions.

Syntax:
lambda arguments: expression

lambda: This keyword is used to define a lambda function.
arguments: A comma-separated list of arguments (just like in regular functions). A lambda function can have multiple arguments but only a single expression.
expression: A single expression whose result is automatically returned. There is no need for an explicit return statement, as the expression is returned by default.

Example:
add = lambda x, y: x + y
result = add(3, 5)
print(result)
Output: 8


Typical Use Cases for Lambda Functions:
1. Used as Arguments to Higher-Order Functions->Lambda functions are commonly used as arguments to higher-order 
functions such as map(), filter(), and sorted(), where a simple, temporary function is needed.

map(): Applies a function to all items in an iterable.
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))  
Output: [1, 4, 9, 16, 25]

2. Used for Simple One-Line Functions->When a simple operation is needed for a short period, and defining a 
full function with def would make the code unnecessarily verbose, lambda functions offer a concise alternative.

Example:
product = lambda a, b: a * b
print(product(4, 5)) 
Output: 20

3. Used with Functions Like reduce()->Lambda functions are also commonly used with the reduce() function, which 
reduces a sequence into a single value by repeatedly applying a function.

Example:
from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)  
Output: 120

4. Used for Conditional Logic in Small Functions->Lambda functions can handle simple conditional logic for 
cases that do not require a full function.

Example:
check_even_odd = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even_odd(4))  
Output: Even
print(check_even_odd(7))  
Output: Odd

Lambda functions in Python are anonymous, small, one-liner functions defined using the lambda keyword.
They are typically used for short, simple operations, especially as arguments to higher-order functions 
(map(), filter(), sorted(), etc.).
They provide a concise way to define functions without the need for a def block.
However, for more complex operations or when reusability is important, regular functions should be preferred.
'''
```

```python
# 9.Explain the purpose and usage of the `map()` function in Python.
```

```python
'''
The map() function in Python is a built-in function used to apply a given function to each item of an iterable 
(like a list, tuple, etc.) and return a map object (which is an iterator) containing the results. 
It allows for the transformation of data from one form to another, applying the specified function to each 
element of the iterable.

Syntax:
map(function, iterable, ...)

function: The function that you want to apply to each item of the iterable(s).
iterable: The iterable (or multiple iterables) whose elements will be passed to the function.
The function is applied to every item of the iterable, and the result is returned as an iterator.

Example1:
# A function to square a number
def square(x):
    return x ** 2

# A list of numbers
numbers = [1, 2, 3, 4, 5]

# Applying the square function to each number using map()
squared_numbers = map(square, numbers)

print(list(squared_numbers))  
Output: [1, 4, 9, 16, 25]

In this example, the map() function applies the square() function to each element in the numbers list and 
returns a map object. We convert it into a list to see the results.

Example2:
When map() is used with multiple iterables, the function must accept multiple arguments. map() will then apply
the function to corresponding elements from each iterable.

# A function to add two numbers
def add(x, y):
    return x + y

# Two lists of numbers
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

summed_numbers = map(add, numbers1, numbers2)
print(list(summed_numbers))  
Output: [5, 7, 9]

In this case, map() takes two lists and adds the corresponding elements together: 1 + 4, 2 + 5, 3 + 6.

Converting the Map Object->The result of map() is an iterator (map object), which is lazily evaluated. 
To see the actual results, we can convert it into a list, tuple, or set:

As a list: result = list(map(function, iterable))
As a tuple:result = tuple(map(function, iterable))
As a set: result = set(map(function, iterable))

Example – Converting to Different Data Structures:
numbers = [1, 2, 3, 4]
# Converting the map object to a list
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers) 
Output: [1, 4, 9, 16]

Use Cases of map():
Applying a transformation to an entire iterable: map() is useful when you need to apply the same operation to every
item in an iterable, such as squaring numbers, converting strings to uppercase, or adding corresponding items 
from multiple lists.
Data Processing Pipelines: It can be used in data processing pipelines to transform data efficiently without 
writing explicit loops.
Advantages of map():
Cleaner Code: Using map() eliminates the need for an explicit loop to iterate over a list and apply a function.
Functional Programming Style: map() supports a functional programming approach, allowing you to focus on applying transformations to data without managing loop mechanics.
Efficiency: map() returns an iterator, which is more memory-efficient than returning a new list immediately, especially when working with large datasets.
Comparison: map() vs for Loop:
You can achieve the same result as map() using a for loop, but map() is more concise and efficient for applying a 
function across an iterable.

Using map():
numbers = [1, 2, 3]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))  
Output: [1, 4, 9]

Using a for loop:
numbers = [1, 2, 3]
squared_numbers = []
for x in numbers:
    squared_numbers.append(x**2)
print(squared_numbers)  
Output: [1, 4, 9]

Summary:
The map() function applies a given function to each item of an iterable and returns an iterator.
It is often used for transforming lists, tuples, or other iterables in a concise and functional programming style.
map() can handle multiple iterables and is useful when performing element-wise operations across them.
The result of map() is an iterator that can be converted into other data structures like lists, tuples, or sets if
needed.
'''
```

```python
# 10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
```

```python
'''
The map(), reduce(), and filter() functions in Python are all used for functional programming, allowing you to 
apply a function to data in a concise manner. However, they serve different purposes and have distinct 
characteristics. 

1. map()
Purpose: Applies a specified function to every item in an iterable (like a list or tuple) and returns an iterator 
containing the results.
Input: A function and one or more iterables.
Output: An iterator of the results (map object).
Usage: Use map() when you want to transform each element in an iterable based on a given function.

Example:
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))
Output: [1, 4, 9, 16, 25]

2. reduce()
Purpose: Applies a specified function cumulatively to the items of an iterable, reducing the iterable to a single 
value.
Input: A function and an iterable (from the functools module).
Output: A single value (result of the reduction).
Usage: Use reduce() when you want to combine items in an iterable into a single result, such as summing all numbers
or multiplying them.

Example:
from functools import reduce
numbers = [1, 2, 3, 4, 5]
sum_result = reduce(lambda x, y: x + y, numbers)
print(sum_result)
Output: 15

3. filter()
Purpose: Filters items from an iterable, returning only those for which the specified function returns True.
Input: A function and an iterable.
Output: An iterator of items that satisfy the filtering condition.
Usage: Use filter() when you want to remove elements from an iterable based on a condition.

Example:
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
Output: [2, 4]

When to Use:
Use map() when you need to apply a function to every element in an iterable and transform it.
Use reduce() when you want to combine elements from an iterable into a single cumulative result.
Use filter() when you need to remove elements from an iterable based on a specific condition.
'''
```
