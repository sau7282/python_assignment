```python
'''
String slicing in Python allows you to extract specific parts (substrings) from 
a string by specifying a range of indices. This feature is useful for working with 
strings in a concise and efficient way. Python strings are immutable, so slicing creates 
new strings rather than modifying the original string.
syntax-> string[start:end:step]

Example:
text = "abcdefghijklmnop"
print(text[0:10:3]) 
output=adgj

String slicing allows for extracting parts of strings with minimal syntax.
Slices are inclusive of the start index and exclusive of the end index.
You can slice with positive or negative indices, and also control the step size.
The step parameter can be used to skip characters or reverse the string.
'''
```

```python
# 3. Describe how to access, modify, and delete elements in a list with examples.
```

```python
'''
1. Accessing Elements in a List->We can access elements in a list by using indexing or slicing.

Access by Index:
Positive indexing starts from 0 (the first element) and moves to the right.
Negative indexing starts from -1 (the last element) and moves to the left.
Example:
my_list = [10, 20, 30, 40, 50]
print(my_list[0])  
Output: 10
print(my_list[2])  
Output: 30
Accessing elements by negative indexing
print(my_list[-1])  
Output: 50
print(my_list[-3])  
Output: 30

Access by Slicing:
Slicing extracts a portion of the list by specifying a start, end, and optionally a step.
The slice includes the element at start but excludes the element at end.
Example:
my_list = [10, 20, 30, 40, 50]
print(my_list[1:4])  
Output: [20, 30, 40]
print(my_list[:3])   
Output: [10, 20, 30]
Slicing with a step
print(my_list[::2])  
Output: [10, 30, 50]

2. Modifying Elements in a List
Lists are mutable, means we can change their elements directly by using indexing or slicing.
Modify by Index:
We can replace an individual element by accessing it via its index and assigning a new value.

Example:
my_list = [10, 20, 30, 40, 50]
Modifying the element at index 1
my_list[1] = 200
print(my_list)  
Output: [10, 200, 30, 40, 50]

Modify by Slicing:
We can modify a group of elements at once by assigning a new list to a slice.

Example:
my_list = [10, 20, 30, 40, 50]
my_list[1:3] = [200, 300]
print(my_list)  
Output: [10, 200, 300, 40, 50]

Changing multiple elements with different-sized lists
my_list[2:4] = [1000, 2000, 3000]
print(my_list)  
Output: [10, 200, 1000, 2000, 3000, 50]

We can append, insert, or extend a list to add new elements.
append(): Adds a single element to the end of the list.
insert(): Inserts an element at a specified position.
extend(): Adds multiple elements from another iterable (e.g., list, tuple).

3. Deleting Elements from a List
We can delete elements from a list using various methods, such as del, pop(), and remove().

Delete by Index:
We can use the del keyword or pop() to remove an element at a specific index.

del: Removes an element (or slice) by index.
pop(): Removes and returns the element at a specific index (or the last element if no index is provided).
Example:
my_list = [10, 20, 30, 40, 50]
Deleting an element by index using del
del my_list[1]
print(my_list)  
Output: [10, 30, 40, 50]

Deleting the last element using pop()
last_element = my_list.pop()
print(last_element)  
Output: 50
print(my_list)       
Output: [10, 30, 40]

Deleting a specific element using pop()
second_element = my_list.pop(1)
print(second_element)  
Output: 30
print(my_list)         
Output: [10, 40]
'''
```

```python
# 4. Compare and contrast tuples and lists with examples.
```

```python
'''
In Python, both tuples and lists are used to store collections of items, but they differ in several key ways,
including mutability, syntax, and use cases. 

1. Mutability->Lists are mutable, meaning you can modify (add, remove, or change) elements after 
the list has been created.
Tuples are immutable, meaning once they are created, their elements cannot be changed.

Example:
my_list = [1, 2, 3]
my_list[0] = 10 
print(my_list)  
Output: [10, 2, 3]

my_tuple = (1, 2, 3)
my_tuple[0] = 10  
This will raise an error: TypeError: 'tuple' object does not support item assignment

2. Syntax->
Lists are defined using square brackets [].
Tuples are defined using parentheses () or simply by separating elements with commas 
(though parentheses make the syntax clearer).

3. Usage Scenarios->
Lists are used when you need a collection of items that can change over time 
(i.e., where elements need to be added, modified, or removed).
Tuples are used when you need a collection that should not change 
(i.e., for fixed collections of items like coordinates or configuration values).

4. Performance->
Tuples are generally faster than lists because they are immutable and have a smaller memory overhead.
Lists, being mutable, require more memory and time for operations like appending or removing elements.

5. Functions/Methods->
Lists have a large number of built-in methods for adding, removing, and modifying elements 
(e.g., append(), remove(), pop()).
Tuples have fewer methods since they are immutable (e.g., count(), index()).
'''
```

```python
# 5.  Describe the key features of sets and provide examples of their use.
```

```python
'''
In Python,sets are an unordered collection of unique elements.
They are primarily used for operations that involve comparing data,removing duplicates,
and performing mathematical set operations like union, intersection, and difference.

Key Features of Sets:
1. Unordered->Sets do not maintain any particular order of elements. 
Unlike lists or tuples, the elements in a set are not indexed, and you cannot access items by index.

Example:
my_set = {3, 1, 4, 2}
print(my_set)  
Output: {1, 2, 3, 4} (the order may vary)

2. Unique Elements->A set automatically removes duplicate elements.
If you add the same element more than once, it will only keep one copy of the element.

Example:
my_set = {1, 2, 2, 3, 4}
print(my_set)  
Output: {1, 2, 3, 4} (duplicates are removed)

3. Mutable (But Elements Must Be Immutable)->While the set itself is mutable
(you can add or remove elements), the elements inside a set must be immutable.
This means you cannot have lists, dictionaries, or other sets as elements, but you can have integers, strings, 
tuples, etc.

Example:
my_set = {1, 2, (3, 4)}  
Valid (tuple is immutable)
my_set = {1, 2, [3, 4]}  
Invalid (list is mutable, so it raises a TypeError)

4. No Indexing or Slicing->Since sets are unordered, you cannot access elements using indexing or 
slicing like you can with lists or tuples. If you need ordered access, you can convert the set to a list.

Example:
my_set = {1, 2, 3}
print(my_set[0])  
This will raise a TypeError because sets do not support indexing
my_list = list(my_set)  
Converting set to a list
print(my_list[0])  
Output: 1 (now indexing is allowed)

5. Set Operations
Sets are particularly useful for mathematical set operations, such as union, intersection, difference, 
and symmetric difference.

Example:
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

Union: Elements in either set
print(set_a | set_b)  
Output: {1, 2, 3, 4, 5, 6}

Intersection: Elements in both sets
print(set_a & set_b)  
Output: {3, 4}

Sets are unordered, mutable collections of unique elements.
Sets support fast membership testing, removing duplicates, and 
performing set operations like union, intersection, and difference.
Key methods include add(), remove(), discard(), and pop().
'''
```

```python
# 7. Describe how to add, modify, and delete items in a dictionary with examples.
```

```python
''' In Python, dictionaries are mutable data structures that store key-value pairs. 
Keys must be unique and immutable, while values can be of any data type. We can easily add, modify, 
or delete items in a dictionary. 

1. Adding Items to a Dictionary->We can add new key-value pairs to a dictionary by assigning a value to a new key. 
If the key already exists, its value will be updated.

Example:
# Creating a dictionary
student = {"name": "Alice", "age": 20}

# Adding a new key-value pair
student["grade"] = "A"

# Output: {'name': 'Alice', 'age': 20, 'grade': 'A'}
print(student)

# Adding multiple key-value pairs using update()
student.update({"major": "Physics", "year": 3})
# Output: {'name': 'Alice', 'age': 20, 'grade': 'A', 'major': 'Physics', 'year': 3}
print(student)

2. Modifying Items in a Dictionary->To modify an item, simply assign a new value to an existing key. 
The dictionary will update the value for that key.

Example:
# Modifying the value of an existing key
student["age"] = 21

# Output: {'name': 'Alice', 'age': 21, 'grade': 'A', 'major': 'Physics', 'year': 3}
print(student)
We can also modify multiple values at once using the update() method:

# Modifying multiple values using update()
student.update({"grade": "A+", "year": 4})

# Output: {'name': 'Alice', 'age': 21, 'grade': 'A+', 'major': 'Physics', 'year': 4}
print(student)

3. Deleting Items from a Dictionary->There are several ways to delete items from a dictionary:

i. Using the del statement->The del statement removes a key-value pair from the dictionary.

Example:
# Deleting an item using del
del student["major"]

# Output: {'name': 'Alice', 'age': 21, 'grade': 'A+', 'year': 4}
print(student)

ii. Using the pop() method->The pop() method removes a key from the dictionary and returns its value. 
If the key doesn’t exist, an error will occur unless a default value is provided.

Example:
# Deleting an item using pop and returning the value
removed_value = student.pop("year")

# Output: 4
print(removed_value)

# Output: {'name': 'Alice', 'age': 21, 'grade': 'A+'}
print(student)

iii. Using the popitem() method->The popitem() method removes and returns the last inserted key-value pair as 
a tuple. 

Example:
# Deleting the last inserted item using popitem()
last_item = student.popitem()

# Output: ('grade', 'A+')
print(last_item)

# Output: {'name': 'Alice', 'age': 21}
print(student)

iv. Using the clear() method->The clear() method removes all items from the dictionary, leaving it empty.

Example:
# Removing all items from the dictionary
student.clear()

# Output: {}
'''
```

```python
'''
In Python, dictionary keys must be immutable because they are used as indices in a hash table 
(the underlying data structure of a dictionary). Immutability ensures that the hash value of a key 
remains constant, which is essential for the efficient retrieval and storage of data in dictionaries. 
If a key's value could change, it would lead to inconsistencies in the dictionary's internal hash map, 
potentially causing errors or making it impossible to retrieve the associated value.

Consistency in Hashing->Python dictionaries use a hashing mechanism to store and retrieve values based on 
their keys. The hash value of a key is used to determine where the key-value pair will be stored in memory. 
Immutable keys ensure that the hash value doesn’t change, so the dictionary can always find the correct 
memory location.

Efficiency->Immutability guarantees that once a key is placed in a dictionary, its hash value is constant. 
This allows for efficient lookup, insertion, and deletion operations, all of which depend on consistent 
hash values for performance optimization.

Avoiding Data Corruption->If a key were mutable and its value changed, the dictionary would not be able to track 
it properly. The key's location in the dictionary would no longer be valid, leading to incorrect retrievals or 
even the loss of the value associated with the key.


Example 1:

# Integer, string, and tuple (which are immutable types) as keys
my_dict = {
    42: "Answer to everything",
    "name": "Alice",
    (10, 20): "Coordinates"
}

# Accessing values using immutable keys
print(my_dict[42])         # Output: Answer to everything
print(my_dict["name"])      # Output: Alice
print(my_dict[(10, 20)])    # Output: Coordinates
Mutable Keys (Invalid): Mutable types like lists, sets, and other dictionaries cannot be used as keys because their values can change, which would lead to unreliable hashing.

Example 2:
my_dict = {}

# Trying to use a mutable type (list) as a key
key = [1, 2, 3]
# This will raise a TypeError: unhashable type: 'list'
my_dict[key] = "This will raise an error"
The above code will fail because lists are mutable, and they cannot be hashed.

'''
```
